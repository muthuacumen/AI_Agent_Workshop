# Setup: Getting Your Anthropic (Claude) API Key

Before we begin building AI Agents, you need an API key to access Claude.
We use individual keys so everyone has dedicated quota and won't hit rate limits.

---

## Step 1 — Get your API key

1. Go to **[console.anthropic.com](https://console.anthropic.com)**
2. Sign in (or create a free account)
3. Click **API Keys** in the left sidebar
4. Click **Create Key**, give it a name (e.g. `workshop`)
5. Copy the key — it starts with `sk-ant-...`
6. Keep it secret — never commit it to GitHub!

---

## Step 2 — Save the key in a `.env` file

Create a file called `.env` in the **project root** (`D:/Projects/AI_Agent_Workshop/`).
Add exactly one line:

```
ANTHROPIC_API_KEY=sk-ant-api03-...
```

The `.env` file is already excluded from Git via `.gitignore` — your key will never be committed.

---

## Step 4 — Verify the key loads correctly

````python
import os
from pathlib import Path
from dotenv import load_dotenv

# Find the project root and load .env
ROOT = Path.cwd()
while not (ROOT / '.env').exists() and ROOT.parent != ROOT:
    ROOT = ROOT.parent

env_path = ROOT / '.env'
if env_path.exists():
    load_dotenv(env_path)
    print(f'Loaded .env from {env_path}')
else:
    print(f'WARNING: .env not found at {env_path}')
    print('Create it with:  ANTHROPIC_API_KEY=sk-ant-...')

api_key = os.getenv('ANTHROPIC_API_KEY')
if api_key and api_key.startswith('sk-ant-'):
    print(f'API key detected: sk-ant-...{api_key[-6:]}')
else:
    print('API key NOT found — check your .env file')
````

## Step 5 — Make a test call to Claude

````python
import anthropic

client = anthropic.Anthropic(api_key=os.getenv('ANTHROPIC_API_KEY'))

message = client.messages.create(
    model='claude-sonnet-4-6',
    max_tokens=256,
    messages=[{
        'role': 'user',
        'content': 'Say "Claude API is working!" and nothing else.'
    }]
)

print(message.content[0].text)
print(f'\nModel used:  {message.model}')
print(f'Stop reason: {message.stop_reason}')
print(f'Input tokens: {message.usage.input_tokens}  |  Output tokens: {message.usage.output_tokens}')
````
